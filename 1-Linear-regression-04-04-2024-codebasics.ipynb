```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model as lin
```

```python
# Set data into dataframe
df = pd.read_csv("Home-prices.csv")
df
```

```python
# Plot the data
%matplotlib inline
plt.scatter(df.Area,df.Prices, marker='+')
plt.xlabel("Area (sq. ft.)")
plt.ylabel("Prices (₹)")
```

```python
# Fit the data into the linear regresiion model
reg=lin.LinearRegression()
reg.fit(df[['Area']],df.Prices)
```

```python
# Predict prices for some new house area data (and save the data alongside the same dataframe in a new column)
d=pd.read_csv('ToPredict2.csv')
p=reg.predict(d)
d['Prices']=p
d
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(y=df.Prices,x=df.Area,marker='+')
plt.plot(df.Area,reg.predict(df[['Area']]),color='red')

```
